```python
#Important features in the dataset.- genres,id,keywords,title,overview,cast,crew.

movies=movies[["movie_id","original_title",'genres','keywords',"overview","cast","crew"]]
movies.head()
```

```python
#changing the format of genres. - ['Action','Adventure','Fantasy','Science Fiction']
import ast   #ast.literal_eval()--evaluates string as literal expression

def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'].apply(convert)
```

```python
movies['genres']=movies['genres'].apply(convert)
movies.head()
```

```python
def convert3(obj):                   
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:              #set counter to 3 for selecting only the first 3 names in the cast features.
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'].apply(convert3)
```

```python
movies['cast']=movies['cast'].apply(convert3)
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
movies.head()
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L

```

```python
movies['crew'].apply(fetch_director)
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
movies.head()
```

```python
movies['overview'].apply(lambda x:x.split())
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])   #for removing spaces between first_name and last_name.
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies.head()
```

```python
#creating a new column tag by concatenating the genres,overview,crew and cast column.

movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
movies.head()
```

```python
#creating a new dataframe which consists only movie_id,original_title and tags column
new_df=movies[['movie_id','original_title','tags']]
new_df.head()
```

```python
new_df['tags']
```

```python
#converting tags column from list to string.
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x)) 
new_df.head()
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
new_df.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):   #words like "loved','loving','love' will be taken as "love" only and will be counted only once.
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
    
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
#Vectorization - converting input data from its raw format(text) into vectors of real numbes.
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
vectors=cv.fit_transform(new_df['tags']).toarray()
vectors
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity(vectors)   #arrays of similarity between the movies
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
new_df.head()
```

```python
def recommend(movie):
    movie_index=new_df[new_df['original_title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].original_title)
```

```python
new_df[new_df['original_title']=='Batman Begins'].index[0]
```

```python
recommend("Batman Begins")
```
